```python
# 字符转义
# 当字符串中包含特殊字符时，需要使用转义字符来表示
# 例如，换行符\n、制表符\t等
# 可以在字符串前添加r或R来表示原始字符串，避免转义字符的解释
s3 = R'hello\nworld'
print(s3)
```

```python
# 读取pdf格式中的表格
import pdfplumber

pdf = pdfplumber.open(R'data\pdf\高考核心词汇1278.pdf')

table = []
# len(pdf.pages)获取全部pdf页数；
for i in range(len(pdf.pages) - 1):
    # 通过循环逐页读取当前页面中的表格；
    page = pdf.pages[i + 1]
    table.extend(page.extract_table())
# 将表格转化为Pandas中的`DataFrame`。
table1_df = pd.DataFrame(table[2:])
table1_df.head(2)
```

```python
# 使用pandas将dataframe保存为csv文件
table1_df.to_csv(R'output/dict.csv', index=False)
```
